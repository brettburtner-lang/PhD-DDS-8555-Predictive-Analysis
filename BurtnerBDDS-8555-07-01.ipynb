# Conceptual question 12.1

### Part a.

##### Equation 12.18 is

<center>$ \frac{1}{\left| \mathrm{C}_{k}^{} \right|}
\sum\limits_{i,i'\in C_k}^{}
\sum\limits_{j = 1}^{p}
(x_{ij} - x_{i'j})^2 = 2\sum\limits_{i\in C_k}^{}
\sum\limits_{j = 1}^{p}
(x_{ij} - \bar{x}_{kj})^2
$
</center>

##### and

<center>
\begin{equation} 
    \begin{split}
        \bar{x}_{kj} = \frac{1}{\left| \mathrm{C}_{k}^{} \right|}
        \sum\limits_{i\in C_k}^{}
        x_{ij} 
    \end{split}
\end{equation}
</center>

##### The left side of equation 12.18 finds the difference between each observation.  The right side finds the difference between each observation and the mean.

##### and

<center>
\begin{equation} 
    \begin{split}
        (x_{ij} - x_{i'j})^2 & =
        ((x_{ij} - \bar{x}_{kj}) - (x_{i'j} - \bar{x}_{kj}))^2 \\
        & = (x_{ij} - \bar{x}_{kj})^2 - 2(x_{ij} - \bar{x}_{kj})(x_{i'j} - \bar{x}_{kj}) + (x_{i'j} -
        \bar{x}_{kj})^2
    \end{split}
\end{equation}
</center>

##### therefore

<center>
\begin{equation}
    \begin{split}
        \frac{1}{\left| \mathrm{C}_{k}^{} \right|}
        \sum\limits_{i,i'\in C_k}^{}
        \sum\limits_{j = 1}^{p}
        (x_{ij} - x_{i'j})^2 
        & =
        \frac{1}{\left| \mathrm{C}_{k}^{} \right|}
        \sum\limits_{i, i'\in C_k}^{}
        \sum\limits_{j = 1}^{p}
        (x_{ij} - \bar{x}_{kj})^2 -
        \frac{1}{\left| \mathrm{C}_{k}^{} \right|}
        \sum\limits_{i,i'\in C_k}^{}
        \sum\limits_{j = 1}^{p}
        2(x_{ij} - \bar{x}_{kj})(x_{i'j} - \bar{x}_{kj}) +
        \frac{1}{\left| \mathrm{C}_{k}^{} \right|}
        \sum\limits_{i,i'\in C_k}^{}
        \sum\limits_{j = 1}^{p}
        (x_{i'j} - \bar{x}_{kj})^2 \\
        & = 
        \frac{\left| \mathrm{C}_{k}^{} \right|}{\left| \mathrm{C}_{k}^{} \right|}
        \sum\limits_{i\in C_k}^{}   
        \sum\limits_{j = 1}^{p}
        (x_{ij} - \bar{x}_{kj})^2 - 
        \frac{2}{\left| \mathrm{C}_{k}^{} \right|}
        \sum\limits_{i,i'\in C_k}^{}
        \sum\limits_{j = 1}^{p}
        (x_{ij} - \bar{x}_{kj})(x_{i'j} - \bar{x}_{kj}) + 
        \frac{\left| \mathrm{C}_{k}^{} \right|}{\left| \mathrm{C}_{k}^{} \right|}
        \sum\limits_{i\in C_k}^{}   
        \sum\limits_{j = 1}^{p}
        (x_{ij} - \bar{x}_{kj})^2 \\
        & = 
        2\sum\limits_{i\in C_k}^{}
        \sum\limits_{j = 1}^{p}
        (x_{ij} - \bar{x}_{kj})^2
    \end{split}
\end{equation}
</center>

##### Equation 12.12 in the text shows the distance between each pair of samples can be related to the difference between each sample and the cluster mean.  The K-means algorithm minimizes the distance to each centroid, thereby minimizing the within-cluster variance.
