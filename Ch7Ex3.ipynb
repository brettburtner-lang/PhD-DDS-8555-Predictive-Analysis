```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2, 2)
y = 1 + x + -2 * (x - 1)**2 * (x >= 1)
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, y)
plt.axvline(x = 1.265, color = 'r', linestyle = '--', linewidth = 0.5)
plt.axhline(y = 2.125, color = 'b', linestyle = '--', linewidth = 0.5)
plt.axvline(x = 1, color = 'g', linestyle = '--', linewidth = 0.5)
plt.show()
```

```python
maxY = np.max(y)
maxX = x[np.argmax(y)]
print('Peak of curve: ', maxY)
print('Value of x at peak of curve: ', maxX)
```
